```python
import pandas as pd

# DataFrames
data1 = {
    'id': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Charlie', 'David']
}

data2 = {
    'id': [3, 4, 5, 6],
    'name': ['Charlie', 'David', 'Eve', 'Frank']
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Difference: Rows in df1 that are not in df2 based on 'id'
difference_df = df1[~df1['id'].isin(df2['id'])]

# Output
print("Rows in df1 but not in df2:")
print(difference_df)
```

```python
import pandas as pd

# Create two dataframes
data1 = {
    'id': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'ahmed', 'mohamr']
}

data2 = {
    'id': [3, 4, 5, 6],
    'name': ['Charlie', 'David', 'Eve', 'Frank']
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Find rows in df1 that are not in df2
difference_df = pd.merge(df1, df2, on=['id','name'], how='left', indicator='df_merge')
#df1_only = difference_df[difference_df['_merge'] == 'left_only'].drop(columns=['_merge'])

print("Rows in df1 but not in df2:")
print(difference_df)
```
