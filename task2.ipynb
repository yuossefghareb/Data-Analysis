```python
import pandas as pd
import numpy as np

```

```python

df = pd.read_csv("Orders2.csv",encoding='unicode_escape',skiprows=4)
df
```

```python
df.info()
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
df.dropna(how='all',inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['City'] = df['City'].str.title()
df
```

```python
df['Branch'] = df['Branch'].str.upper()
df
```

```python
df['Email'] = df['Email'].str.lower()
df
```

```python
df['Phone Number'] = df['Phone Number'].str.replace('Tel: ','')
df
```

```python
df['Discount'] = df['Discount'].fillna(0)
df
```

```python
df['Total Price'] = df['Quantity'] * df['Unit Price']
df
```

```python
df.isnull().sum()
```

```python
df['Cost'] = df['Total Price'] - df['Discount']
df
```

```python
df.insert(9,'Customer Full Name',df['Customer First Name'] + ' '+ df['Customer Last Name'] )
```

```python
df
```

```python
newcolumn = df['Branch'].str[:2]
df.insert(5,'Code',newcolumn)
df
```

 assign specific names or labels to the bins created from the data

```python
df['Price Category']=pd.qcut(df['Unit Price'],q=3,labels=['low cost','mid range','high cost'])
```

```python
df[['Unit Price','Price Category']]
```

map categories to integer indices,

```python
df['Discount_Category_Index']=pd.qcut(df['Discount'],q=3).cat.codes
df
```

using aggfunc before bins 

```python
category_sales = df.groupby('Category')['Total Price'].sum()
category_sales
```

```python
category_sales_bins = pd.cut(category_sales, bins=3, labels=['Low Sales', 'Medium Sales', 'High Sales'])
```

```python
category_sales_bins
```

```python
customer_sales = df.groupby('SalesPerson ID')['Total Price'].sum()
```

```python
customer_sales_bins = pd.cut(customer_sales, bins=3, labels=['Low Sales', 'Medium Sales', 'High Sales'])

```
