```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Orders2.csv",encoding='unicode_escape',skiprows=4)
df.head()
```

```python
df.info()
```

```python
df.dropna(how='all',inplace=True)
```

```python
df.isnull().sum()
```

```python
df['City'] = df['City'].str.title()
df['Branch'] = df['Branch'].str.upper()
df['Email'] = df['Email'].str.lower()
df['Phone Number'] = df['Phone Number'].str.replace('Tel: ','')
df['Discount'] = df['Discount'].fillna(0)
df['Total Price'] = df['Quantity'] * df['Unit Price']
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Cost'] = df['Total Price'] - df['Discount']
```

```python
df.insert(9,'Customer Full Name',df['Customer First Name'] + ' '+ df['Customer Last Name'] )
```
